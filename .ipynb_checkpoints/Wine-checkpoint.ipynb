```python
from sklearn.datasets  import load_wine as wine
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
import mglearn
```

```python
wine_dataset = wine()
```

```python
wine_dataset.DESCR
```

```python
print(wine_dataset["feature_names"])
```

```python
print(wine_dataset["target"].shape)
```

```python
print("wines type data - {}".format(wine_dataset["data"].shape))
```

Gets the Sample dataSet Count for Each target 

```python
print("Sample counts per class:\n{}".format(
  {wine for wine in zip(wine_dataset.target_names, np.bincount(wine_dataset.target))}))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(wine_dataset["data"], wine_dataset["target"], random_state = 0)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
wine_dataframe = pd.DataFrame(X_train, columns = wine_dataset["feature_names"])
wine_dataframe
```

```python
pd.plotting.scatter_matrix(wine_dataframe, c=Y_train, figsize=(40, 40), marker='o',
                            hist_kwds={'bins': 20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, Y_train)
```

```python
input = [[13.64,3.10,2.56,15.2,116.0,2.70,3.03,0.17,1.66,5.10,0.96,3.36,845.0]]
prediction = knn.predict(input)
print("Predicted target name: {}".format(
wine_dataset['target_names'][prediction]))
    
```

```python
print("Test set score: {:.2f}".format(np.mean(prediction == Y_test)))
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(X_train, Y_train)
```

```python
print("Score For Training Data {:.2f}".format(lr.score(X_train, Y_train)))
print("Score For Test Data {:.2f}".format(lr.score(X_test, Y_test)))
```

- decrease Alpha improves generalization

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 1).fit(X_train, Y_train)
```

```python
print("training data : {:.2f}".format(ridge.score(X_train,Y_train)))
print(ridge.score(X_test,Y_test))
```
