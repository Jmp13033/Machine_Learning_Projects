# Support Vector Machine
- classification problem
- nearby data points are called support vectors 
- Hyperplane : -> a decision boundary that differentiates the two classes in SVM
- decision maximizing the margin between the points and the slope 
- line with higher margin is better 
- if data is hard to fit consider creating a new dimension z = x2 + y2

# kernal 
- Kernels are a way to solve non-linear problems with the help of linear classifiers
- These kernel functions also help in giving decision boundaries for higher dimensions.

# C 
- regularization in SVM model 

# gamma
- high or low gamma depends on individual points
- high regulularization makes training data fit well
- low regularization may get some errors
- high gamma: considers data points that are very near to the decision boundary
- Low gamma : considers points farther away from the decision boundary as well
