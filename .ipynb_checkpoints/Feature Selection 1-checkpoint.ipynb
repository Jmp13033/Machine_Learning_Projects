```python
country_code = pd.read_excel("Country-Code.xlsx")

country_code
```

```python
# merge on, how?
final_df = pd.merge(df,country_code, on = "Country Code", how="left")
```

```python
final_df
```

```python
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
values = final_df.Country.value_counts().values
```

```python
plt.pie(values[:3], labels = country_names[:3], autopct="%1.2f%%")
```

Maximum Records are From India USA and United Kingdom

```python
final_df.columns
```

```python
Aggregated_Ratings = final_df.groupby(["Aggregate rating",'Rating color', "Rating text" ]).size().reset_index().rename(columns = {0:"Rating Count"})
```

```python
Aggregated_Ratings
```

```python
sns.barplot(x="Aggregate rating", data=Aggregated_Ratings, y = "Rating Count")
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=Aggregated_Ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
## countplot categorical variables
```

```python
sns.countplot(x= "Rating color", data=Aggregated_Ratings, palette=['blue','red','orange','yellow','green','green'])
```

```python
country_color_ratings = final_df[final_df["Rating color"] == "White"].groupby(["Aggregate rating","Country"]).size().reset_index().rename( columns ={0:"Count"})

country_color_ratings
```
