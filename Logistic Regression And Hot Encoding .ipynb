```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

import plotly.express as px



```

```python
df = pd.read_csv("HR_comma_sep.csv")
df.columns
```

```python
df
```

```python
df.left.value_counts()
```

```python
dummies = pd.get_dummies(data = df, columns = ["Department", "salary"])
dummies.drop("left", axis = 1)
merged_df = pd.concat([df, dummies], axis = 1)
final = merged_df.drop(["left","Department","Department_marketing", "salary_high", "salary"], axis = 1)

final






```

```python
# final 

```

```python
X = final
y = df["left"]
X





```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=40)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
model = LogisticRegression(max_iter=100).fit(X_train, y_train)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test, y_test))
```

```python
y_prediction =model.predict(X_test)
y_prediction[:100]




```
