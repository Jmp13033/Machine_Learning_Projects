 # KNN Classifier
 - for the kNN we need a two dimensional plane... 
 - XY
 - subplots(1,4)
 - mglearn.plots.plot_2d_separator
 - mglearn.discrete_scatter

```python
fig, axes = plt.subplots(1, 4, figsize =(15, 3))
for n_neighbors, ax in zip([1, 3, 9, 7], axes):
    clf = KNeighborsClassifier(n_neighbors= n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.7, ax=ax, alpha=.7) 
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")


```

# KNN Regressor

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
X, y = mglearn.datasets.make_wave(n_samples= 40)
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=0)
```

 - the R2 score. The R2 score, also known as the coefficient of determination, is a measure    of goodness of a prediction for a regression model closer to one the better model. 

```python
# build the model 
# fit the model with training data
# predict the test 
# score the test 
# remember X and Y test variab;es go together IE input(X) and output 
reg = KNeighborsRegressor(n_neighbors=3)
fit = reg.fit(X_train, Y_train)
# predict the x_test set 
prediction = reg.predict(X_test)
print("Test set R^2: {:.2}".format(reg.score(X_test, Y_test)))
print("Test Score : {:.2f}".format(reg.score(X_test, Y_test)))







 
```

- the score is 0.83, which indicates a relatively good model fit.
